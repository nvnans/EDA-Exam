```python
# Handle outliers in 'Salary' using capping
# Determine a reasonable upper limit (e.g., based on the 95th percentile or visual inspection)
# From the describe output, the max is 1,000,000, while 75% is around 160,000.
# Let's cap at a value slightly above the 75th percentile, e.g., 300,000.
upper_limit = 300000
df['Salary'] = df['Salary'].clip(upper=upper_limit)

print(f"\nStatistik deskriptif untuk 'Salary' setelah capping (upper limit = {upper_limit}):")
display(df['Salary'].describe())

# Visualisasi 'Salary' setelah capping (Box Plot)
plt.figure(figsize=(5, 5))
sns.boxplot(data=df, y='Salary')
plt.title('Box Plot of Salary After Capping')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
```

```python
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)

# Reset indeks dan tambahkan 1 untuk membuat indeks dimulai dari 1
df = df.reset_index(drop=True)
df.index = df.index + 1

# Tampilkan beberapa baris pertama DataFrame untuk memastikan data berhasil dimuat
display(df.head())

# Menampilkan korelasi antara Usia dan Pendapatan
correlation = df['Age'].corr(df['Salary'])
print(f"Korelasi antara Age dan Salary: {correlation}")

# Hitung koefisien determinasi (R-squared)
r_squared = correlation**2
print(f"Koefisien determinasi (R-squared) antara Age dan Salary: {r_squared}")
```

```python
import statsmodels.formula.api as smf
import pandas as pd

file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)

# Reset indeks dan tambahkan 1 untuk membuat indeks dimulai dari 1
df = df.reset_index(drop=True)
df.index = df.index + 1

# Definisikan dan sesuaikan model regresi Pendapatan vs Usia
model_usia_pendapatan = smf.ols('Salary ~ Age', data=df).fit()

# Mengambil koefisien dari model regresi Pendapatan vs Usia
intercept_usia_pendapatan = model_usia_pendapatan.params['Intercept']
usia_coef_usia_pendapatan = model_usia_pendapatan.params['Age']

# Mengambil Jumlah Kuadrat Residual (SSR) dari model baru
ssr_usia_pendapatan = model_usia_pendapatan.ssr

print(f"Jumlah Kuadrat Residual (SSR) untuk model Salary vs Age: {ssr_usia_pendapatan:.4f}")
print("Persamaan Garis Regresi (Salary vs Age):")
print(f"Salary_prediksi = ({intercept_usia_pendapatan:.4f}) + ({usia_coef_usia_pendapatan:.4f}) * Age")
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

shapiro_test = stats.shapiro(df['Salary'])
print(f"Hasil Uji Shapiro-Wilk: statistik = {shapiro_test.statistic:.4f}, p-value = {shapiro_test.pvalue:.4f}")

alpha = 0.05
if shapiro_test.pvalue > alpha:
    print("Data kemungkinan berdistribusi normal (gagal menolak H)")
else:
    print("Data tidak berdistribusi normal (menolak H)")

plt.figure(figsize=(8, 6))
sns.histplot(df['Salary'], kde=True)
plt.title('Histogram of Salary with KDE')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```
